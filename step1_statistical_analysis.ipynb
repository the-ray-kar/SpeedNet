```python
import numpy as np
```

```python
def load_label_data():
        labels = []
        with open("../sp_data/train.txt", 'r') as file:
             for line in file:
                line.replace("\n","")
                labels.append(float(line))

        return labels
```

```python
speed_data = np.array(load_label_data())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(speed_data)
#Speed varies from high to low. The speed variation is not homogenous during the journey
#The video then needs to be split train and test to cover all speeds
```

```python
hist = plt.hist(speed_data,bins=10,density=True)
print(max(speed_data),min(speed_data))
```
